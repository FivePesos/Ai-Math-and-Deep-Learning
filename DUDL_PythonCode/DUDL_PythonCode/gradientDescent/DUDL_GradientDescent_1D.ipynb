```python
# import all necessary modules
import numpy as np
import matplotlib.pyplot as plt

import matplotlib_inline.backend_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')
```

# Gradient descent in 1D

```python
# function (as a function)
def fx(x):
  return 3*x**2 - 3*x + 4

# derivative function
def deriv(x):
  return 6*x - 3
```

```python
# plot the function and its derivative

# define a range for x
x = np.linspace(-2,2,2001)

print(x)

# plotting
plt.plot(x,fx(x), x,deriv(x))
plt.xlim(x[[0,-1]])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['f(x)','d(fx)'])
plt.show()
```

```python
# random starting point
localmin = np.random.choice(x,1)
print(localmin)

# learning parameters
learning_rate = .01
training_epochs = 100

# run through training
for i in range(training_epochs):
  grad = deriv(localmin) #Compute the derivative at current guess
  localmin = localmin - learning_rate*grad #W = W-N * derivative; W= guess, N= learning rate

localmin
```

```python
# plot the results

plt.plot(x,fx(x), x,deriv(x))
plt.plot(localmin,deriv(localmin),'ro')
plt.plot(localmin,fx(localmin),'ro')

plt.xlim(x[[0,-1]])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['f(x)','df','f(x) min'])
plt.title('Empirical local minimum: %s'%localmin[0])
plt.show()
```

# Store the model parameters and outputs on each iteration

```python
# random starting point
localmin = np.random.choice(x,1)
print(localm)
# learning parameters
learning_rate = .01
training_epochs = 100

# run through training and store all the results
modelparams = np.zeros((training_epochs,2))
for i in range(training_epochs):
  grad = deriv(localmin)
  localmin = localmin - learning_rate*grad
  modelparams[i,0] = localmin[0]
  modelparams[i,1] = grad[0]
```

```python
# plot the gradient over iterations

fig,ax = plt.subplots(1,2,figsize=(12,4))

for i in range(2):
  ax[i].plot(modelparams[:,i],'o-')
  ax[i].set_xlabel('Iteration')
  ax[i].set_title(f'Final estimated minimum: {localmin[0]:.5f}')

ax[0].set_ylabel('Local minimum')
ax[1].set_ylabel('Derivative')

plt.show()
```

```python
# random starting point
localmin = np.random.choice(x, 1)

# learning parameters
learning_rate = 0.01
threshold = 0.01   # stop if slope is smaller than this
max_epochs = 1000 # safety cap (to prevent infinite loops)

# store results
modelparams = []

for i in range(max_epochs):
    grad = deriv(localmin)

    # stop if slope is smaller than threshold
    if abs(grad) < threshold:
        print(i)
        break

    localmin = localmin - learning_rate * grad
    modelparams.append([localmin[0], grad[0]])

modelparams = np.array(modelparams)  # convert to array for plotting
```

```python
# plot the gradient over iterations

fig,ax = plt.subplots(1,2,figsize=(12,4))

for i in range(2):
  ax[i].plot(modelparams[:,i],'o-')
  ax[i].set_xlabel('Iteration')
  ax[i].set_title(f'Final estimated minimum: {localmin[0]:.5f}')

ax[0].set_ylabel('Local minimum')
ax[1].set_ylabel('Derivative')

plt.show()
```
