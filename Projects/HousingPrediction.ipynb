```python
#Option 1
#new data set without price column. Note: all categorical data were removed already
X = df.drop(['price'],axis=1)
#dependent value/target variable
y = df['price']
```

```python
#Option 2
#Place the data into an X array that contains independent variables (explanatory variable) and y (target variable)
#by enumerating all the columns to be included in the array

#independent values X array does not include the price
X = df[['Area Income','House Age','Number of Rooms','Number of Bedrooms','Area Population']]
#dependent value/target variable y Price
y = df['Price']
```

```python
X
```

```python
y
```

```python
#split the normalized data using 60:40 as the ratio; but note that in model training 80:20 is recommended
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=10)
```

```python
print("Training set", X_train.shape)
print("Testing set", X_test.shape)
```

```python
#Build the Classifier: import the Linear Regression library
from sklearn.linear_model import LinearRegression
```

```python
#create A Linear Regression model
lm = LinearRegression()
```

```python
#Train the data model
lm.fit(X_train,y_train)
```

```python
#get the intercept
intercept = lm.intercept_

#display the Intercept
print (f"Intercept : {intercept: .2f}")
```

```python
#get the slope of each independent variable (features)

coeff_df= pd.DataFrame(lm.coef_, X.columns, columns=['Coefficient'])
```

```python
#show the coefficient
print (coeff_df)
```

```python
#Predict house price using the training data
y_pred_train = lm.predict(X_train)
y_pred_train
```

```python
#predict the house price from the first training data
print (y_pred_train[0])
```

```python
#Predict house price using the test data
y_pred_test = lm.predict(X_test)
y_pred_test
```

```python
#Compute and display the predicted against the actual
y_pred_df = pd.DataFrame({'Actual Price': y_test, 'Predicted Price': y_pred_test, 'Difference': y_test-y_pred_test})
y_pred_df[0:20]
```

```python
#visualize the difference between actual and prediction results.

import matplotlib.pyplot as plt

#plt.figure(figsize=(10,10))
plt.scatter(y_train, y_pred_train, label= 'Training data',color='b', alpha=0.7)
plt.plot([min(y_train), max(y_train)], [min(y_train), max(y_train)], 'r', label='Regression line')


plt.legend()
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Price vs Predicted Price')
plt.show()
```

```python
#r2 measures how close are the data fitted to the regression line
# r_square answers the question: Is the model is good or bad?

from sklearn.metrics import r2_score

score = r2_score(y_train,y_pred_train) # portion of the variance that can be explained by the training features
print (f"Score: {score*100: .2f}%") #coeff of determination
```
