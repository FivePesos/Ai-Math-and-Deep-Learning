```python
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt 
```

# Using Numpy

```python
# "Manually" in numpy

# The list of numbers 
z = [1,2,3]

# Compute the softmax 
num = np.exp(z) #calculate the exponential function to the power of z elements
den = np.sum(np.exp(z)) #calculate the sum of e^z

sigma = num/den #divide

print(f"{num}\n")
print(f"{den}\n")
print(f"{sigma}\n")
print(np.sum(sigma)) # sum of sigma
```

```python
# Repeat with some random integers
z = np.random.randint(-5, high = 15, size= 25)
print(f"{z} \n")

num = np.exp(z)
den = np.sum(np.exp(z))
sigma = num/den #divide

#compare
plt.plot(z,sigma, "ko")
plt.xlabel("Original number (z)")
plt.ylabel("SoftMaxified $\\sigma$")
plt.yscale("log")
plt.title("$\\sum\\sigma$ = %g" %np.sum(sigma))
plt.show()
```

# Using pytorch

```python
#slightly more involved using torch.nn

#create an instance of the softmax activation class
softfun = nn.Softmax(dim = 0)

#then apply the data to that function
sigmaT = softfun(torch.Tensor(z))

#results
print(sigmaT)
```
